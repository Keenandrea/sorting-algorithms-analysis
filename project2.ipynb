# **Selection Sort**

---


An instance of the Selection Sort algorithm implemented in Python 3. Iterate over the unsorted array, keeping track of the minimum value as we go so that when we get to the end of the array, we know which element is the minimum. Swap the minimum element element and the first element in the unsorted array. The first element is now considered sorted. We repeat until the rest of the array is sorted.

No matter what, Selection Sort has a time complexity of O(n²), for both worst case and best case performance, which ranks it among the worst sorting algorithms.

```
def selection_sort(lst):        
    for i in range(len(lst)):
        minimum = i
        
        for j in range(i + 1, len(lst)):
            # Select the smallest val
            if lst[j] < lst[minimum]:
                minimum = j

        # Place the smallest value it at the front 
        # of the sorted end of the array
        lst[minimum], lst[i] = lst[i], lst[minimum]
            
    return lst
```

```
selection_sort(lst=[9,4,3,7,6,1,8,2,10,5])
```

# **Insertion Sort**

---


An instance of the Insertion Sort algorithm implemented in Python 3. Considering the first element to be sorted and the rest to be unsorted. Compare the first element with the second. If the second element is less than the first element, insert the element in the correct position of the sorted part, otherwise leave it as is. Repeat this procedure until all the elements are sorted. 

Insertion Sort works best with a small number of elements. The worst case complexity of Bubble Sort is O(n²), on par with Bubble Sort, however Insertion Sort is often considered better than Bubble Sort because of what occurs in each pass. Bubble Sort swaps through all remaining unsorted values, whereas Insertion Sort swaps a value up into already-sorted values. 


```
def insertion_sort(lst):
   for i in range(1,len(lst)):

      #element compare
      current = lst[i]

      #comparing the current element with the sorted portion and swapping
      while i>0 and lst[i-1]>current:
        lst[i] = lst[i-1]
        i = i-1
        lst[i] = current
       
      return lst
```

```
insertion_sort(lst=[9,4,3,7,6,1,8,2,10,5])
```

# **Bubble Sort**

---

An instance of Bubble Sort algorithm implemented in Python 3. Bubble sort steps through the given list and compares adjacent pairs of elements. The elements are swapped if they are in the wrong order. The pass through the unsorted portion of the list is repeated until the list is sorted. 

The trouble of Bubble Sort is it's habit to repeatedly passes through the unsorted part of the list, which marks it with worst case complexity of O(n²).

```
def bubble_sort(lst):
    def swap(i, j):
        lst[i], lst[j] = lst[j], lst[i]

    n = len(lst)
    swapped = True
    
    x = -1
    while swapped:
        swapped = False
        x = x + 1
        for i in range(1, n-x):
            if lst[i - 1] > lst[i]:
                swap(i - 1, i)
                swapped = True
                    
    return lst
```

Bubble Sort implementation to show it's in working condition before running tests.

```
bubble_sort(lst=[9,4,3,7,6,1,8,2,10,5])
```

# **Bubble Sort: Swaps & Comparisons Counted**

---

An instance of Bubble Sort algorithm with swaps and comparisons counted implemented in Python 3. Bubble sort steps through the given list and compares adjacent pairs of elements, incrementing a counter at each instance. The elements are swapped if they are in the wrong order, and a counter is incremented. The pass through the unsorted portion of the list is repeated until the list is sorted.

The trouble of Bubble Sort is it's habit to repeatedly passes through the unsorted part of the list, which marks it with worst case complexity of O(n²).

```
def bubble_sort_stats(lst):
    cmpcount, swapcount = 0, 0
    
    for j in range(len(lst)):
        
        for i in range(1, len(lst)-j):
            cmpcount += 1
            
            if lst[i-1] > lst[i]:
                swapcount += 1
                lst[i-1], lst[i] = lst[i], lst[i-1]
                
    return lst, cmpcount, swapcount
```

```
bubble_sort_stats(lst=[9,4,3,7,6,1,8,2,10,5])
```

# **Quick Sort**

---

An instance of Quick Sort algorithm implemented in Python. Quick Sort is a divide-and-conquer algorithm. First, Quick Sort selects an element which we call a pivot from the list. Next, we move all elements that are smaller than the pivot to the left of the pivot, and all the elements that are larger than the pivot to the right of the pivot, which is a method called a partition operation. Finally, we apply the aforementioned two steps seperatly to each of the sub-arrays created by the partition.

In most standard implementations, Quick Sort performs significantly faster than the other divide-and-conquer algorithm, Merge Sort, which will be presented next. Quick Sort also rarely reaches its worst case complexity of O(n²).

```
def partition(xs, start, end):
    follower = leader = start
    while leader < end:
        if xs[leader] <= xs[end]:
            xs[follower], xs[leader] = xs[leader], xs[follower]
            follower += 1
        leader += 1
    xs[follower], xs[end] = xs[end], xs[follower]
    return follower

def _quicksort(xs, start, end):
    if start >= end:
        return
    p = partition(xs, start, end)
    _quicksort(xs, start, p-1)
    _quicksort(xs, p+1, end)
    
def quicksort(xs):
    _quicksort(xs, 0, len(xs)-1)
    return xs
```

```
quicksort(xs=[2,4,5,6,3,4])
```

# **Merge Sort**

---

An instance of Merge Sort algorithm implemented in Python 3. Merge Sort is a textbook example of a divide-and-conquer algorithm. It consists of two parts: the divide part, and the conquer part. It continuously divides the unsorted list until we have N sublists, where each sublist has one element that is unsorted, and where N is the number of elements in the original list. Then it repeatedly merges, or conquers the sublists together two at a time to produce new sorted sublists until all the elements have been fully merged into one sorted list.  

Merge Sort and Quick Sort contend for the title of best divide-and-conquer algorithm. As previously stated Quick Sort, in most standard implementations, performs significantly faster than Merge Sort. However, in the worst case, Merge Sort does fewer comparisions than Quick Sort does in it's average case. The worst case complexity of Merge Sort is O(nlogn), which is the same complexity as Quick Sort's best case. Further analysis will be presented below. 

```
def merge_sort(arr):
    # last array split
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    # perform merge_sort recursively on both halves
    left, right = merge_sort(arr[:mid]), merge_sort(arr[mid:])

    # merge each side together
    return merge(left, right, arr.copy())


def merge(left, right, merged):

    left_cursor, right_cursor = 0, 0
    while left_cursor < len(left) and right_cursor < len(right):
      
        # sort each one and place into the result
        if left[left_cursor] <= right[right_cursor]:
            merged[left_cursor+right_cursor]=left[left_cursor]
            left_cursor += 1
        else:
            merged[left_cursor + right_cursor] = right[right_cursor]
            right_cursor += 1
            
    for left_cursor in range(left_cursor, len(left)):
        merged[left_cursor + right_cursor] = left[left_cursor]
        
    for right_cursor in range(right_cursor, len(right)):
        merged[left_cursor + right_cursor] = right[right_cursor]

    return merged
```

Dig deeper into the timing of Selection Sort and a pattern emerges. Let's lay the groundwork. The time complexity of Selection Sort is O(n²). n² across the board. That means it's quadratic, involving the second and no higher power of an unknown quantity or variable. Now for the pattern I touched on. Under our conditions, Selection Sort takes approximately *n^2/10^7* seconds to sort* n* values. For a fairly small value of *n*, let's say *n* = 1000, the running time of Selection Sort is about 1000^2/10^7 = 0.1 second, which is where we get our timing value of ~0.123. The list then grows by a factor of 10, but the running time increases 100 times. When *n*=10,000, the running time of our Selection Sort is about 10,000^2/10^7, which is ~10 seconds, not shy off the mark of our ~11.47 timing. What if *n*=1,000,000? Just one zero more than our maximum list length and we'd be looking at 11.6 days of sorting. We ought to consider ourselves lucky.

Next sorting algorithm under the microscope is Insertion Sort. This one has the same worst time and average time complexities as Selection Sort with the saving grace of a big-Omega *n* best case. 

Our Insertion Sort implementation kept within a range between 4 and 8 seconds for each of the nine given lists. Why? Well, let's see. Our Insertion Sort uses a *while* loop to loop over the indices of the list. It *inserts* on the elements at indices 1,2,3,..., *n* - 1. Suppose that upon every call to *insert*, the value being inserted is less than every element in the subarray to its left. That being so, when we call *insert* the first time, k = 1. The second time, k = 2. And so on, until k = n - 1. Thusly, the total time spent inserting into sorted subarrays is going to be c*1 + c*2 + c*3 + ... c *(n-1) = c *(1 +2 + 3 + ...)*. That sum is an arithmetic series. Using the formula, we get that the total time spent inserting is *c*(n -1 + 1)((n - 1)/2) = cn^2/2 - cn/2. Running time: big-theta n^2. The value to consider her is *c*. Whatever constant is being multiplied into our implementation is giving it it's tight-bound timing.

Merge Sort is interesting because the initial *merge* function runs in big-theta of n time when merging n elements, and then gets down to big-theta of nlog-base2n time. It's in the magic of the divide-and-conquer. 

Merge Sort works just like the recursion trees we learned to solve in class. It divides the problem, say n, into subproblems of size n/2 and recursively sorts tow subarrays of size (n/2)/2, or n/4. And so on. Then the level is merged. So if the second level is n/4, then the merging will be 4 x cn/4, which equals cn. As the subproblems get smaller, the number of subproblems doubles at each level of the recursion, but the merging time halves. It's a fascinating algorithm. Quick too.
